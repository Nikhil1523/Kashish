```python
# Matplotlib
# It is a popular python library used to create all kinds of charts and graphs
# 1. Easy to use
# 2. Wide variety of plots
# 3. Highly cutomizable
# 4. Integrate with other libraries
```

```python
import matplotlib.pyplot as plt
```

```python
# Line PLot
# Used to show trend over time
# Trend - Direction over a period of time
plt.figure(figsize = (8,5)) #sets the size if the graph (width, height)
month = ['January','February','March','April']
sales = [3000,2000,5000,4000]
expenses = [1000,5000,3000,4500]

#plt.plot(month,sales)

# Marker marks the datapoints and ms is for marker size, mfc is marker color, mec is for marker edge color
plt.plot(month,sales,marker='*',ms=15,mfc='red',mec='black',linestyle='-.',color='red',linewidth=5.5)
plt.plot(month,expenses,marker='s',ms=15,mfc='blue',mec='black',linestyle='-.',color='green',linewidth=5.5)
plt.xlabel('Month')
plt.ylabel('Sales and Expenses')
plt.title('Sales of First-Quarter')
plt.grid(axis = 'y')
plt.show()
```

```python
# Bar Plots
categories = ['A','B','C']
values = [20,30,40]
plt.bar(categories,values,color='orange',width = 0.6)
plt.show()
#plt.barh() #for creating horizontal bars, height parameter can also be used just like width for normal bar
#plt.bar_label(plt.gca().containers[0])
# gca : fetching the current graph area
```

```python
# scatter plot
# study hours and marks =  +ve relationship
# watching tv hrs increase and marks decrease  => -ve relation
# shoe size and intelligence = 0 or on relation

study_hr = [1,2,3,4,5]
marks = [50, 60, 80, 87, 90]

plt.scatter(study_hr, marks, color ='red')
plt.show()
```

```python
# Pie Chart
# Show Proportion
a = [10,20,30,40]
labels = ['Mango','Apple','Kiwi','Dates']

plt.pie(a, labels = labels, autopct ='%.1f%%')
plt.show()
```

```python
# Seaborn : a high level visualization library built on matplotlib, makes graphs look more attractive, easier to create
# and effective for statistical analysis.

import seaborn as sns
x = [10,20,30,40]
y = [20,30,40,50]
sns.lineplot(x=x,y=y)
plt.show()
```

```python
x = ['Samsung','Vivo','Oppo']
y = [10000,25000,15000]
color = ['red','blue','green']
sns.barplot(x=x,y=y,palette=color)
plt.show()
```

```python
age = [10,20,30,20,20,20,30,40,40,50,60,10,10]
sns.histplot(age, bins = 3)
plt.show()
```

```python
sns.set_style('darkgrid')
temp = [10,20,30,40]
sales = [1000,2000,3000,4000]
sns.scatterplot(x=temp,y=sales)
plt.show()
```
